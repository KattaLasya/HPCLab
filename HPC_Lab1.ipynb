Method 1: Simple Serial Parameter Sweep

```python
import random

def mc_pi(N):
    inside = 0
    for _ in range(N):
        x, y = random.random(), random.random()
        if x*x + y*y <= 1:
            inside += 1
    return 4.0 * inside / N
```

```python
samples = [1000, 5000, 10000, 50000, 100000]

for i, N in enumerate(samples):
    pi_est = mc_pi(N)
    print(f"Task {i}: N={N}, pi≈{pi_est}")
```

Method 2: Parallel Execution

```python
from multiprocessing import Pool
import random

def mc_pi(N):
    inside = 0
    for _ in range(N):
        x, y = random.random(), random.random()
        if x*x + y*y <= 1:
            inside += 1
    return N, 4.0 * inside / N

samples = [1000, 5000, 10000, 50000, 100000]

with Pool(processes=4) as p:
    results = p.map(mc_pi, samples)

for N, pi_est in results:
    print(f"N={N}, pi≈{pi_est}")
```
