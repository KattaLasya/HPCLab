Import libraries & initialize array

```python
import numpy as np
from numba import njit, prange

# Initialize a large array
N = 10_000_000
arr = np.ones(N, dtype=np.int64)

print("Array initialized with", N, "elements")
```

Serial sum (baseline)

```python
# Serial sum for reference
serial_sum = np.sum(arr)

print("Serial Sum:", serial_sum)
```

Incorrect parallel attempt (race condition example)

```python
@njit(parallel=True)
def parallel_sum_race(a):
    total = 0
    for i in prange(len(a)):
        total += a[i]   # Race condition (no reduction)
    return total

race_sum = parallel_sum_race(arr)

print("Parallel Sum (Race Condition):", race_sum)
```

Correct parallel reduction (race-free)

```python
@njit(parallel=True)
def parallel_sum_reduction(a):
    total = 0
    for i in prange(len(a)):
        total += a[i]   # Safe reduction
    return total

parallel_sum = parallel_sum_reduction(arr)

print("Parallel Sum (Reduction):", parallel_sum)
```

Final comparison

```python
print("Serial Sum:           ", serial_sum)
print("Race Condition Sum:   ", race_sum)
print("Correct Reduction Sum:", parallel_sum)
```
